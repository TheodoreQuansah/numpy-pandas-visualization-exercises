Use matplotlib to plot the following equation: y = x**2 - x + 2
    
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.
    

```python
def equation(x):
    
    return x ** 2 - x + 2
```

```python
x_val = list(range(1, 20))
```

```python
y_val = [equation(x) for x in x_val]
```

```python
plt.plot(x_val, y_val, label="y = x^2 - x + 2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Equation")
plt.grid(True)
plt.legend()
plt.annotate('(0, 0)', xy=(0, 0))
```

4) Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 

Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10, 12))
x = np.arange(0, 50)

#1st subplot
y1 = np.sqrt(x)
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='turquoise', label='y = sqrt(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('1st Graph')

#2nd subplot
y2 = x**3
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='yellow', label='y = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2nd Graph')

#3rd subplot
y3 = np.power(2, x)
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='blue', label='y = 2^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('3rd Graph')

#4th subplot
y4 = [1 / (val + 1) for val in x]
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='orange', label='y = 1 / (x + 1)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('4th Graph')

plt.suptitle('Y Graphs', fontsize=50)
plt.legend()
plt.show()
```
