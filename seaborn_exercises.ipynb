Use seaborn's load_dataset function to load the iris database to answer the following questions:

3)Would it be reasonable to predict species based on sepal width and sepal length? 

For this, you'll visualize two numeric columns through the lense of a categorical column.

```python
x = 'sepal_length'
y = 'sepal_width'
colors = {'setosa': 'purple', 'versicolor': 'orange', 'virginica': 'green'}
sns.scatterplot(data = iris, x = x, y = y, hue = 'species', marker = 'x', palette = colors)
plt.show()
```

4) Which features would be best used to predict species?

```python
sns.boxplot(y = 'petal_width', x = 'species', data = iris)
```

4) Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from sqlalchemy import create_engine, text
from env import get_connection
```

```python
def get_db_url(database):
    return get_connection(database)
```

```python
url = get_db_url('chipotle')
query = text('SELECT * FROM orders')
engine = create_engine(url)
```

```python
orders = pd.read_sql(query, engine.connect())
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '').astype('float')
```

```python
chicken_bowl = orders[orders.item_name == 'Chicken Bowl']
chicken_bowl.sort_values('quantity')
```

```python
best_sellers = orders.groupby('item_name').quantity.sum().nlargest(4)
best_sellers
```

```python
revenue = orders[orders.item_name.isin(best_sellers.index)].groupby('item_name').item_price.sum()
revenue
```

```python
revenue = revenue.sort_values().reset_index()
```

```python
plt.figure(figsize = (10, 5))
sns.barplot(y = 'item_name', x = 'item_price', data = revenue)
plt.title('Top selling items and revenue')
plt.xlabel("Revenue")
plt.suptitle("Top 4 selling Items")
plt.show()
```
