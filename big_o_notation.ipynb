```python
#random data range
x = np.arange(1, 30)
```

```python
#curves based on Big O notations
y1 = 0 * x + 1

y2 = np.log(x)

y3 = x

y4 = x * np.log(x)

y5 = x ** 2

y6 = 2 ** x

y7 = [np.math.factorial(n) for n in x]

y8 = x ** x

plt.figure(figsize = (18,8))

#Plotting the curves with labels
plt.plot(x, y1, color='silver', label=r'y1=$O(1)$', alpha = 1)

plt.plot(x, y2, color='Olive', label=r'y2=$O(\log n)$', alpha = 1)

plt.plot(x, y3, color='Navy', label=r'y3=$O(n)$', alpha = 1)

plt.plot(x, y4, color='Teal', label=r'y4=$O(n \log n)$', alpha = 1)

plt.plot(x, y5, color='Maroon', label=r'y5=$O(n^2)$', alpha = 1)

plt.plot(x, y6, color='Lime', label=r'y6=$O(2^n)$', alpha = 1)

plt.plot(x, y7, color='Magenta', label=r'y7=$O(n!)$', alpha = 1)

plt.plot(x, y8, color='Cyan', label=r'y8=$O(n^n)$', alpha = 1)

# Adding labels and a legend 
plt.xlabel('Elements', fontsize=12)

plt.ylabel('Operations', fontsize=12)

plt.annotate('LaTex', xy= (1.5,10))

plt.title('Big O Notation', fontsize=20)

plt.legend(loc = 'upper left', fontsize = 10)

plt.xlim(1, 10)

plt.ylim(0, 40)


plt.show()
```
